```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob

# -------------------------
# PARAMETERS
# -------------------------
waveform_dir = "waveform_raw"
output_dir = "waveform_plots"
os.makedirs(output_dir, exist_ok=True)

# -------------------------
# PLOTTING LOOP
# -------------------------
waveform_files = sorted(glob.glob(os.path.join(waveform_dir, "waveform_*.txt")))

for wf_file in waveform_files:
    idx = os.path.basename(wf_file).split("_")[1].split(".")[0]
    truth_file = os.path.join(waveform_dir, f"truth_{idx}.txt")

    # Load waveform
    data = np.loadtxt(wf_file, skiprows=1)
    time, amplitude = data[:, 0], data[:, 1]

    # Load truth t0s and amplitudes
    t0s, amps = [], []
    with open(truth_file, "r") as f:
        for line in f:
            if line.startswith("Signal Index") or line.startswith("Number"):
                continue
            parts = line.strip().split()
            if len(parts) == 3:
                t0s.append(float(parts[1]))
                amps.append(float(parts[2]))

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(time, amplitude, color='gray', label="Waveform")
   # plt.scatter(t0s, amps, color='blue', edgecolors='k', label="True Signals", s=60)
    plt.title(f"Waveform #{idx}")
    plt.xlabel("Time (ns)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, f"waveform_{idx}.png"))
    plt.close()

print(f"✅ Plotted {len(waveform_files)} waveforms to '{output_dir}'")
```
