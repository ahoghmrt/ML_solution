```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import load_model

# Load dataset
data = np.load("ml_training_data/training_data.npz")
X = data["waveforms"]
y_true = data["labels"]
counts_true = data["counts"]
time = data["time"]

# Load model
model = load_model("ml_training_data/signal_extraction_model.keras")

# Normalize input
X_mean = X.mean(axis=1, keepdims=True)
X_std = X.std(axis=1, keepdims=True)
X_norm = (X - X_mean) / (X_std + 1e-8)

# Predict
pred_signal, pred_count = model.predict(X_norm)
max_signals = y_true.shape[1]
pred_signal = pred_signal.reshape((-1, max_signals, 3))
pred_count_labels = np.argmax(pred_count, axis=1)

# Evaluate
mae_t0_list, mae_amp_list, mae_pres_list = [], [], []
rmse_t0_list, rmse_amp_list, rmse_pres_list = [], [], []

for i in range(max_signals):
    mae_t0_list.append(mean_absolute_error(y_true[:, i, 0], pred_signal[:, i, 0]))
    rmse_t0_list.append(np.sqrt(mean_squared_error(y_true[:, i, 0], pred_signal[:, i, 0])))
    mae_amp_list.append(mean_absolute_error(y_true[:, i, 1], pred_signal[:, i, 1]))
    rmse_amp_list.append(np.sqrt(mean_squared_error(y_true[:, i, 1], pred_signal[:, i, 1])))
    mae_pres_list.append(mean_absolute_error(y_true[:, i, 2], pred_signal[:, i, 2]))
    rmse_pres_list.append(np.sqrt(mean_squared_error(y_true[:, i, 2], pred_signal[:, i, 2])))

# Plot
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
signals = np.arange(1, max_signals + 1)

axs[0].bar(signals, mae_t0_list, label="MAE")
axs[0].bar(signals, rmse_t0_list, bottom=mae_t0_list, label="RMSE")
axs[0].set_title("t0 Errors per Signal")
axs[0].set_xlabel("Signal Index")
axs[0].set_ylabel("Error (ns)")
axs[0].legend()

axs[1].bar(signals, mae_amp_list, label="MAE")
axs[1].bar(signals, rmse_amp_list, bottom=mae_amp_list, label="RMSE")
axs[1].set_title("Amplitude Errors per Signal")
axs[1].set_xlabel("Signal Index")
axs[1].set_ylabel("Error (ADC units)")
axs[1].legend()

axs[2].bar(signals, mae_pres_list, label="MAE")
axs[2].bar(signals, rmse_pres_list, bottom=mae_pres_list, label="RMSE")
axs[2].set_title("Presence Errors per Signal")
axs[2].set_xlabel("Signal Index")
axs[2].set_ylabel("Error")
axs[2].legend()

plt.tight_layout()
plt.savefig("ml_training_data/model_performance.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
import os

# Load dataset
data = np.load("ml_training_data/training_data.npz")
X = data["waveforms"]
y_true = data["labels"]  # shape: (n_samples, max_signals, 3)
counts_true = data["counts"]
time = data["time"]

# Load model and normalize input
model = load_model("ml_training_data/signal_extraction_model.keras")
X_mean = X.mean(axis=1, keepdims=True)
X_std = X.std(axis=1, keepdims=True)
X_norm = (X - X_mean) / (X_std + 1e-8)

# Predict
pred_signal, pred_count = model.predict(X_norm)
max_signals = y_true.shape[1]
pred_signal = pred_signal.reshape((-1, max_signals, 3))
pred_count_labels = np.argmax(pred_count, axis=1)

# Choose sample indices to visualize
sample_indices = [10, 11, 12, 13, 14]  # Feel free to change this list
output_dir = "ml_training_data/prediction_samples"
os.makedirs(output_dir, exist_ok=True)

for idx in sample_indices:
    waveform = X[idx]
    true_t0s = y_true[idx, :, 0]
    true_amps = y_true[idx, :, 1]
    true_pres = y_true[idx, :, 2]

    pred_t0s = pred_signal[idx, :, 0]
    pred_amps = pred_signal[idx, :, 1]
    pred_pres = pred_signal[idx, :, 2]

    plt.figure(figsize=(12, 6))
    plt.plot(time, waveform, label="Waveform", color='black')

    # Plot true signals
    for i in range(max_signals):
        if true_pres[i] > 0.5:
            plt.plot(true_t0s[i], true_amps[i], 'go', markersize=10, label='True Signal' if i == 0 else "")

    # Plot predicted signals
    for i in range(max_signals):
        if pred_pres[i] > 0.5:
            plt.plot(pred_t0s[i], pred_amps[i], 'rx', markersize=10, label='Predicted Signal' if i == 0 else "")

    plt.title(f"Sample {idx} | True Count: {int(counts_true[idx])} | Predicted Count: {int(pred_count_labels[idx])}")
    plt.xlabel("Time (ns)")
    plt.ylabel("ADC Amplitude")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"{output_dir}/waveform_prediction_{idx}.png")
    plt.show()
```
