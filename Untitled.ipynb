```python
# Cell 1: Load dataset
import numpy as np
import matplotlib.pyplot as plt

data = np.load("ml_training_data/training_data.npz")
X = data["waveforms"]                      # shape: (n_samples, time_bins)
y = data["labels"]                         # shape: (n_samples, max_signals, 3)
counts = data["counts"]                    # shape: (n_samples,)
time = data["time"]                        # shape: (time_bins,)

# Show data structure
print(f"✅ Loaded dataset:")
print(f"   ▶ Waveforms     → {X.shape} = (n_samples, time_bins)")
print(f"   ▶ Labels        → {y.shape} = (n_samples, max_signals, [t0, amplitude, presence])")
print(f"   ▶ Signal Counts → {counts.shape} = (n_samples,), unique values: {np.unique(counts)}")
print(f"   ▶ Time Axis     → {time.shape} = (time_bins,)")

# Optional: print first label array for a sample
print("\nSample [0] Labels:")
for i in range(y.shape[1]):
    t0, amp, presence = y[6, i]
    print(f"  Slot {i+1}: t₀ = {t0:.2f}, A = {amp:.2f}, Presence = {presence:.0f}")

# Optional quick view of first waveform
plt.plot(time, X[6])
plt.title("Sample Waveform [0]")
plt.xlabel("Time (ns)")
plt.ylabel("ADC Amplitude")
plt.grid(True)
plt.show()

```

```python
# Cell 2: Plot example waveform with truth signal positions
sample_id = 68  # change to try other samples

plt.figure(figsize=(12, 5))
plt.plot(time, X[sample_id], label="Waveform", linewidth=1.5)

for i in range(y.shape[1]):
    t0, amp = y[sample_id, i]
    if amp > 0:
        plt.plot(t0, amp, 'go', label=f'Truth Signal {i+1}' if i == 0 else "")

plt.xlabel("Time (ns)")
plt.ylabel("Amplitude")
plt.title(f"Sample #{sample_id} - Waveform with Truth Signal Positions")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Cell 3: Quick check of all truth amplitudes and t0s
truth_t0s = y[:, :, 0].flatten()
truth_amps = y[:, :, 1].flatten()

# Filter out zero-padded ones
truth_t0s = truth_t0s[truth_amps > 0]
truth_amps = truth_amps[truth_amps > 0]

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(truth_t0s, bins=30)
plt.xlabel("t0 (ns)")
plt.title("Distribution of True t0 Values")

plt.subplot(1, 2, 2)
plt.hist(truth_amps, bins=30)
plt.xlabel("Amplitude")
plt.title("Distribution of True Amplitudes")

plt.tight_layout()
plt.show()
```
