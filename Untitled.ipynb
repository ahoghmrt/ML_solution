```python
# plot_signals_from_training_data.py

import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Load Dataset
# -------------------------------
data = np.load("ml_training_data/training_data_signals.npz")
X = data["waveforms"]
y = data["labels"]         # shape: (n_samples, max_signals, 2)
time = data["time"]

print(f"✅ Loaded {X.shape[0]} waveforms")

# -------------------------------
# Plot Function
# -------------------------------
def plot_waveform_with_truth(sample_id):
    waveform = X[sample_id]
    truth = y[sample_id]

    plt.figure(figsize=(12, 5))
    plt.plot(time, waveform, label="Waveform", linewidth=1.5, color='black')

    for i, (t0, amp) in enumerate(truth):
        if amp > 0:
            plt.plot(t0, amp, 'go', markersize=8, label=f"Signal {i+1}" if i == 0 else "")

    plt.title(f"Waveform #{sample_id} with Ground Truth Signals")
    plt.xlabel("Time (ns)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# -------------------------------
# Try Example Sample
# -------------------------------
example_id = 23  # Change this to try other examples
plot_waveform_with_truth(example_id)
```
